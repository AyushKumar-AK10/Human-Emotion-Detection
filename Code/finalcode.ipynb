```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
from keras.models import load_model
model=load_model('model1.h5')
```

```python
import cv2
import numpy as np

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# To capture video from webcam
cap = cv2.VideoCapture(0)

# Create a window
cv2.namedWindow('Human Emotion Detection ©Ayush Productions', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Human Emotion Detection ©Ayush Productions', 500, 500)

while True:
    # Check if window is closed
    if cv2.getWindowProperty('Human Emotion Detection ©Ayush Productions', cv2.WND_PROP_VISIBLE) < 1:  
        break

    # Read the frame
    _, img = cap.read()

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect the faces
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        # Extract face
        face = gray[y:y+h, x:x+w]
        # Resize face to match model's expected input size
        resized_gray_face = cv2.resize(face, (48, 48))
        yhat = model.predict(np.expand_dims(resized_gray_face/255, 0))
        output = np.argmax(yhat)
        label = ""
        if(output==0):
            label = 'Angry'
        elif(output==1):
            label = 'Disgusted'
        elif(output==2):
            label = 'Feared'
        elif(output==3):
            label = 'Happy'
        elif(output==4):
            label = 'Neutral'
        elif(output==5):
            label = 'Sad'
        elif(output==6):
            label = 'Surprised'
        # Draw the rectangle around each face
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        # Put label on the rectangle
        cv2.putText(img, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,0,255), 2)

    # Display
    cv2.imshow('Human Emotion Detection ©Ayush Productions', img)

    # Stop if escape key is pressed
    k = cv2.waitKey(30) & 0xff
    if k==27:
        break
        
# Release the VideoCapture object
cap.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
